# Dataset and Preprocessing
In this section we describe all the datasets used in the creation of the visualizations for our data story and display the first 5 entries in these datasets.

## Dataset 1 : Greenhouse gas emission in the Netherlands

This dataset includes emissions of the gases carbon dioxide, methane, and nitrogen in the Netherlands from 1990 up until 2017 and is used in figure 4, 5 and 6. The data is ordered by emission source, such as pharmaceutical industry, households, agriculture and fishery. The emissions themselves are measured in million kilograms, and are continuous ratio values. This dataset can be used to calculate increase in greenhouse gas emissions and compare against datasets of the emissions of other countries. The dataset was taken from <a href='https://www.kaggle.com/datasets/janheindejong/greenhouse-gas-emissions-in-the-netherlands?select=IPCC_emissions.csv'> Kaggle </a> and needed to be seperated by semicolon to be deemed usable. 



```python
import pandas as pd

df_greenhouse = pd.read_csv("../datasets/IPCC_emissions.csv", sep=';')
df_greenhouse.head()
```

## Dataset 2 : Statistics per Country

This dataset shows several different statistics from nearly every county in de United Nations in 2017, and was used in figure 2. Some of these 50 statistics are population, GDP, agriculture economy, internet use, and emission estimates. The variables used in this datastory are the population in thousands and the % of employment in agriculture, which are both discrete ratio values. The dataset was preprocessed from its original dataset (country_profile_variables.csv) to include only a few relevant European countries before it was used (EUonly_country_profile_variables.csv) and can be found on <a href='https://www.kaggle.com/datasets/sudalairajkumar/undata-country-profiles?select=kiva_country_profile_variables.csv'>Kaggle</a>. 

```python
eu_country_profile = pd.read_csv("../datasets/EUonly_country_profile_variables.csv")
eu_country_profile.head()
```

## Dataset 3 : Arable Land in the Europe

This dataset contains data on the percentage of arable land in different European countries and is used to create figure 3. Since the dataset is used to compare against dataset 2, we only used the values that describe the year 2017. The values in this column are the percentages of arable land in different countries, which is a discrete ratio value. The dataset was preprocessed from its original dataset (arable_land_dataset.csv) to include only a few relevant European countries before it was used (EUonly_arable_land_dataset.csv) and can be found on <a href='https://data.worldbank.org/indicator/AG.LND.ARBL.ZS?view=map&year=2020'>The World Bank</a>.



```python
df_arable = pd.read_csv("../datasets/EUonly_arable_land_dataset.csv")
df_arable.head()
```

## Dataset 4 : Global Air Pollution

This dataset contains geolocated data on greenhouse gas emissions and AQI of air pollution of large cities in nearly all countries and was used in figure 1. Some of the attributes are the overall AQI value and category, as well as the AQI value and category for the gases N2O, CO and the Ozone. The dataset can be used to compare air pollution between different countries. The values are measured, which makes them continuous ratio values. The dataset shown below was not processed, but was cleaned before usage by grouping the data by country and aggregating the temperature values. The dataset can be found on <a href='https://www.kaggle.com/datasets/hasibalmuzdadid/global-air-pollution-dataset'>Kaggle</a>.

```python
df_pollution = pd.read_csv("../datasets/global_air_pollution_dataset.csv")
df_pollution.head()
```

## Dataset 5 : Agricultural companies in the Netherlands
This dataset includes data on the agricultural companies in the Netherlands over the period of 1851-2022, and was used to create figure 6. The variables in the dataset that were used in this story include the total number of agricultural companies, the number of companies that hold cattle and the number of companies that hold pigs. These values are all discrete and ratio values. The dataset was preprocessed by appending 'JJ00' to the years in the 'Perioden' column, as that simplified the comparison between this dataset and dataset 1. Furthermore, the columns describing companies with cattle and pigs were combined while making figure 6 using a sum of the values. The dataset can be found on the website of the <a href='https://www.cbs.nl/nl-nl/cijfers/detail/71904ned#'>Centraal Bureau voor Statistiek</a>.

```python
df_agr_comp = pd.read_csv('../datasets/table__71904ned.csv')
df_agr_comp.head()
```

## Dataset 6 : Import and Export of the Netherlands
This dataset describes the import income and export income of goods from the Netherlands to countries inside or outside Europe in the years 2021 and 2022 and was used in the creation of figure 7. The variables 'export income', 'meat and meatproducts', 'dairyproducts and eggs', 'grain and grainproducts', 'vegetables and fruit' and 'natural fertilizers and raw mineral products' are used to make figure 6. These variables describe products from the agricultural sector, are measured in million euros and are discrete ratio values. The variable 'export' is chosen, because we want to show what the income is that is produced by argiculture.  The dataset was not preprocessed and can be found on the website of the <a href='https://www.cbs.nl/nl-nl/cijfers/detail/83926NED?dl=6C56B#'>Centraal Bureau voor Statistiek</a>.
